```python
import pandas as pd

# Create a series with 5 elements
s = pd.Series([50, 20, 40, 10, 30], index=['e', 'c', 'd', 'a', 'b'])
print("Original Series:\n", s)

# Sort by index
print("\nSorted by Index:\n", s.sort_index())

# Sort by values
print("\nSorted by Values:\n", s.sort_values())
```

```python
# Create a series with duplicate values
s2 = pd.Series([10, 20, 20, 30, 10, 40])
print("\nSeries with duplicates:\n", s2)

# Rank using 'first' method
rank_first = s2.rank(method='first')

# Rank using 'max' method
rank_max = s2.rank(method='max')

print("\nRank (method='first'):\n", rank_first)
print("\nRank (method='max'):\n", rank_max)

# Minimum and Maximum ranks
print("\nMin rank (first):", rank_first.min())
print("Max rank (first):", rank_first.max())

print("\nMin rank (max):", rank_max.min())
print("Max rank (max):", rank_max.max())
```

```python
# Create a sample series
s3 = pd.Series([15, 5, 25, 10], index=['A', 'B', 'C', 'D'])
print("\nSeries:\n", s3)

# Index of minimum value
print("\nIndex of minimum value:", s3.idxmin())

# Index of maximum value
print("Index of maximum value:", s3.idxmax())
```
