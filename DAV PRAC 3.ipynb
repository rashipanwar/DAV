```python
import pandas as pd
import numpy as np

# Create DataFrame with 50 rows and 3 columns
df = pd.DataFrame(np.random.randint(1, 100, size=(50, 3)), columns=['A', 'B', 'C'])

# Insert 10% null values randomly
for _ in range(int(df.size * 0.1)):
    i = np.random.randint(0, df.shape[0])
    j = np.random.randint(0, df.shape[1])
    df.iat[i, j] = np.nan

print("Original DataFrame:\n", df)
```

```python
missing_counts = df.isnull().sum()
print("Missing values per column:\n", missing_counts)
```

```python
df = df.loc[:, df.isnull().sum() <= 5]
print("After dropping columns with >5 nulls:\n", df)
```

```python
row_sum = df.sum(axis=1)
max_row = row_sum.idxmax()
df = df.drop(max_row)
print("After dropping row with max sum:\n", df)
```

```python
first_col = df.columns[0]
df = df.sort_values(by=first_col)
print("Sorted DataFrame by first column:\n", df)
```

```python
df = df.drop_duplicates(subset=first_col)
print("After removing duplicates from first column:\n", df)
```

```python
cols = df.columns
if len(cols) >= 3:
    corr = df[cols[0]].corr(df[cols[1]])
    cov = df[cols[1]].cov(df[cols[2]])
    print("Correlation (Column 1 & 2):", corr)
    print("Covariance (Column 2 & 3):", cov)
else:
    print("Not enough columns left after dropping.")
```

```python
second_col = df.columns[1]
df['Binned'] = pd.cut(df[second_col], bins=5)
print("DataFrame with Bins:\n", df[[second_col, 'Binned']])
```
