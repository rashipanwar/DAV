```python
import pandas as pd

df1 = pd.read_excel("workshop1.xlsx")
df2 = pd.read_excel("workshop2.xlsx")

print("Workshop 1:\n", df1)
print("\nWorkshop 2:\n", df2)
```

```python
#(A)
common = pd.merge(df1, df2, on='Name')
print("Students in both workshops:\n", common['Name'].unique())
```

```python
#(B)
names1 = set(df1['Name'])
names2 = set(df2['Name'])
single = names1.symmetric_difference(names2)
print("Students in only one workshop:\n", single)
```

```python
#(C)
merged = pd.concat([df1, df2])
print("Merged DataFrame:\n", merged)
print("Total records:", len(merged))
```

```python
#(D)
merged = pd.concat([df1, df2])
# Set multi-index (Name + Date)
merged.set_index(['Name', 'Date'], inplace=True)
# Descriptive statistics
print("\nDescriptive Statistics:\n", merged.describe())
```
