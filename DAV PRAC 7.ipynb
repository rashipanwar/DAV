```python
import pandas as pd

data = {
    'FamilyName': ['Shah', 'Vats', 'Vats', 'Kumar', 'Vats', 'Kumar', 'Shah', 'Shah', 'Kumar', 'Vats'],
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male'],
    'MonthlyIncome': [44000, 65000, 43150, 66500, 255000, 103000, 55000, 112400, 81030, 71900]
}

df = pd.DataFrame(data)
print(df)
```

```python
# (a)
family_income = df.groupby('FamilyName')['MonthlyIncome'].sum()
print("Family-wise Gross Monthly Income:\n", family_income)
```

```python
# (b)
income_stats = df.groupby('FamilyName')['MonthlyIncome'].agg(['max', 'min'])
print("\nHighest and Lowest Income:\n", income_stats)
```

```python
# (c)
less_income = df[df['MonthlyIncome'] < 80000]
print("\nMembers earning less than 80000:\n", less_income)
```

```python
# (d)
female_data = df[df['Gender'] == 'Female']

total_females = female_data.shape[0]
avg_female_income = female_data['MonthlyIncome'].mean()

print("\nTotal Females:", total_females)
print("Average Female Income:", avg_female_income)
```

```python
# (e)
avg_income = df['MonthlyIncome'].mean()

df_filtered = df[df['MonthlyIncome'] >= avg_income]

print("\nData after deleting rows with income less than average:\n", df_filtered)
```
