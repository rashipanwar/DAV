```python
#calculate mean
mean = np.mean(arr1,axis=1)
print(mean)
```

```python
#calculate standard deviation
std = np.std(arr1,axis=1)
print(std)
```

```python
#calculate variance
var = np.var(arr1,axis=1)
print(var)
```

```python
# G
m=int(input("enter m:"))
n=int(input("enter n:"))

# creating a 2D array with integer elements
arr = np.arange(m*n).reshape(m,n)
print(arr)
```

```python
arr_swap= arr.copy()
arr_swap[[0,1]]=arr_swap[[1,0]]
print(arr_swap)
```

```python
arr_rev=arr_swap.copy()
arr_rev[:,1]=arr_rev[::-1,1]
print(arr_rev)
```
