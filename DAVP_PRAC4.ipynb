```python
import pandas as pd
```

```python
# Data for Workshop 1
data1 = {
    'Name': ['Aman', 'Riya', 'Sohan', 'Kiran', 'Aman', 'Sohan', 'Kiran', 'Riya'],
    'Date': ['2026-01-01', '2026-01-01', '2026-01-02', '2026-01-02', '2026-01-03',
             '2026-01-03', '2026-01-04', '2026-01-04'],
    'Duration': [30, 40, 50, 30, 40, 50, 30, 40]
}

df1 = pd.DataFrame(data1)


# Data for Workshop 2
data2 = {
    'Name': ['Aman', 'Neha', 'Riya', 'Sohan', 'Kiran', 'Neha', 'Aman', 'Kiran'],
    'Date': ['2026-02-01', '2026-02-01', '2026-02-02', '2026-02-02', '2026-02-03',
             '2026-02-03', '2026-02-04', '2026-02-04'],
    'Duration': [40, 30, 50, 40, 30, 50, 40, 30]
}

df2 = pd.DataFrame(data2)
```

```python
# Convert dataframes into excel files
df1.to_excel("Workshop1.xlsx",index=False)
df2.to_excel("Workshop2.xlsx",index=False)
```

```python
df1 = pd.read_excel("Workshop1.xlsx")
df2 = pd.read_excel("Workshop2.xlsx")
```

```python
# Take unique names
names1 = df1[['Name']].drop_duplicates()
names2 = df2[['Name']].drop_duplicates()
# Perform inner join merge
both_students = pd.merge(names1,names2,on='Name',how='inner')
print("Students who attended both workshops ")
print(both_students)
```

```python
# use outer join
single_students = pd.merge(names1,names2,on='Name',how='outer',indicator=True)
single_students = single_students[single_students['_merge']!='both']
print('Students who attended only one workshop:')
print(single_students)
```

```python
# Merge two dataframes row-wise and find total total records
merged_df = pd.concat([df1,df2])
total_records = len(merged_df)
print("Total number of records: " , total_records)
```

```python
merged_df = pd.concat([df1,df2])
merged_df.set_index(['Name','Date'],inplace = True)
stats = merged_df.describe()
print("descriptive statistics")
print(stats)
```
